# Contents of Dataset

Fixed acidity: refers to the quantity of tetratic acid in the wine measured in grams per cubic meter. Here, the values are all numeric.

Volatile acidity:The quantity of acetic acid in the wine measured in g/dm3 is referred to as volatile acidity. Here, the values are all numeric.

Citric acid: The quantity of acid in the wine, which is likewise measured in g/dm3, is referred to as this characteristic. Here, the values are all numeric.

Residual Sugar:The quantity of sugar left in the wine after fermentation is referred to as residual sugar. The unit of measurement is g/dm3. Here, the values are all numeric.

Free sulphar dioxide : This attribute relates to the quantity of SO2 in free form. It's also expressed in g/dm3. Here, the values are all numeric.

Total sulfur dioxide: This attribute relates to the wine's total level of Sulphar Dioxyde. Here, the values are all numeric.

Density: This characteristic refers to the wine's density, which is measured in g/dm3. Here, the values are all numeric.

The pH value of the wine is referred to as the pH characteristic. The value ranges from 0 to 14.0, with 14.0 denoting very high acidity and 14 denoting basic acidity. Here, the values are all numeric.

Sulphates: The quantity of K2SO4 in the wine is referred to as this characteristic. It's also expressed in g/dm3. Here, the values are all numeric.

This trait relates to the amount of alcohol in the wine. Here, the values are all numeric.
